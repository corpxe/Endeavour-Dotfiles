```python
# What is Statistics?
# Stats is the way of collecting, organizing, and interpreting data to understand things better and make informed data-drive decisions.

# Need for Stats :- 
    # To make meaningful conclusions
    # To make better decisions
    # To find relationships
    # To test various ideas/assumptions


# Application of Stats
    # 1) HealthCare
    # 2) Cricket
    # 3) Machine Learning
    # 4) E-Commerce
    # 5) Business

# What does Stats do?

        # collecting data - Surveys, experiments, ebservations
        # Summarizes the data - Using terms like average, percentages
        # Find the patterns - identify the trends, relationships, predictions
        # helps us in decision making - used in business,sports etc.

# Types of Stats 

    # 1) Descriptive Statistics
    # 2) Inferential Statistics


# Descriptive Stats

    # Summarizes and describes the characteristics of a data. 
    # Used to analyze and present the data in a meaningful way,
    # Ex. Mean, Median and Mode

# Inferential Stats

    # Draws/ Derives conclusions or make predictions about a population upon a sample.
    # Used to generalize the findings beyond the data
    # Ex. Hypothesis Testing, CI (Confidence Interval), etc.
```
