```python
# What is Statistics?
# Stats is the way of collecting, organizing, and interpreting data to understand things better and make informed data-drive decisions.

# Need for Stats :- 
    # To make meaningful conclusions
    # To make better decisions
    # To find relationships
    # To test various ideas/assumptions


# Application of Stats
    # 1) HealthCare
    # 2) Cricket
    # 3) Machine Learning
    # 4) E-Commerce
    # 5) Business

# What does Stats do?

        # collecting data - Surveys, experiments, ebservations
        # Summarizes the data - Using terms like average, percentages
        # Find the patterns - identify the trends, relationships, predictions
        # helps us in decision making - used in business,sports etc.

# Types of Stats 

    # 1) Descriptive Statistics
    # 2) Inferential Statistics


# Descriptive Stats

    # Summarizes and describes the characteristics of a data. 
    # Used to analyze and present the data in a meaningful way,
    # Ex. Mean, Median and Mode

# Inferential Stats

    # Draws/ Derives conclusions or make predictions about a population upon a sample.
    # Used to generalize the findings beyond the data
    # Ex. Hypothesis Testing, CI (Confidence Interval), etc.
```

```python
# Descriptive statistics

# Measures of central tendency:

    # Mean(Average) : Sum of all values/ Count of values
    # Median : Middle value after sorting
    # Mode : Most frequency/ observes value
```

```python
import numpy as np
import statistics as stats
```

```python
# Using numpy
data = [5,2,7,8,5,6,9,1,4,5,3]
mean = np.mean(data)
print(mean)

#Using Stats
ans = stats.mean(data)
print(ans)
```

```python
# Median
# odd count of values - middle value(median)
# even count of values - average of two middle values(median)
```

```python
# Using numpy
data = [5,2,7,8,5,6,9,1,4,5,3,1]
median = np.median(data)
print(median)
```

```python
# Using stats
data = [5,2,7,8,5,6,9,1,4,5,3,1]
median = stats.median(data)
print(median)
```

```python
#Mode
data = [5,2,7,4,4,8,5,6,9,1,4,4,4,5,3]
b = stats.mode(data)
print(b)
```

```python
data = [1,2,3,4,2,5,6,8,9,8]
n = stats.mode(data)
print(n)

#Multimode
e = stats.multimode(data)
print(e)
```

```python
# Measures of Variaton / Despersion
    # Tells about the spread in data.


# Range
# Percentagle
# Quartile 
# Variance
# Standard Deviation
# outliers
# corelation
```

```python
data1 = [68,69,70,71,72]
data2 = [10,40,70,90,140]

m1 = np.mean(data1)
m2 = np.mean(data2)

print(m1, m2)
```

```python
# Range - The difference between the maximum and the minimum data point/valules.

data = [25,30,10,90,65,78,34,56,78,32,92]

max_value = max(data)
min_value = min(data)
range_value = max_value - min_value
print(range_value)
```

```python
range = np.ptp(data) #ptp -> Peak to peak
print(range)
```

```python
# 1 - 100
# 100 - 1 = 99
# min data point is 99 units away from the max data point.
```

```python
# Quartiles

    # Steps ->
        # 1) Find median
        # 2) Take data before median and calculate Average = Q1 -> 25%
        # 3) Take data after median and calculate Average = Q3 -> 75%
        # 4) Q2 -> 50%

        # Q1 is the first 25 percentile of the data
        # Q2 is the middle 50 percentile of the data
        # Q3 is the last 25 percentile of the data

    # Interquartile range(IQR)
        # IQR -> Q3 - Q1
        # 7.5 - 2.5 => 5
```
