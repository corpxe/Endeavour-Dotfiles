```python
arr1 = np.random.randint(1,10,5)
arr2 = np.random.randint(1,10,5)
arr3 = np.random.randint(1,10,5)

print(arr1,"\n",arr2,"\n",arr3)

arr4 = np.subtract(arr3,arr2)
print(arr4)

arr5 = np.multiply(arr1,2)
print(arr5)

print("\n")
print(np.cov(arr1,arr4))
print(np.corrcoef(arr1,arr5))
```

```python
#3a 
df = pd.DataFrame(np.random.randn(50,3),columns=['A','B','C'])
for _ in range(int(0.1*df.size)):
    i = np.random.randint(0,50)
    j = np.random.randint(0,3)
    df.iat[i,j] = np.nan

print(df)
```

```python
print("Missing Values:\n",df.isnull().sum())

df = df.dropna(axis = 1, thresh=45)
print(df)
```

```python
if df.shape[1] > 0:
    idmax = df.sum(axis=1).idxmax()
    df = df.drop(idmax)

print(df)
```

```python
if "A" in df.columns:
    df = df.sort_values(by='A')
    df = df.drop_duplicates(subset='A')

if 'A' in df.columns and 'B' in df.columns:
    print(df['A'].corr(df['B']))


if 'B' in df.columns:
    df['B_bins'] = pd.cut(df['B'], bins=5)
    print(df.head())
```
